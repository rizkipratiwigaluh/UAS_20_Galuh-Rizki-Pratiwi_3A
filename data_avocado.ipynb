```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("avocado.csv", usecols=["average_price","total_volume","type","year","geography"])
```

```python
f = plt.figure(figsize=(20,20))

f.add_subplot(1,2,1)
sns.countplot(df['year'])

f.add_subplot(1,2,2)
plt.boxplot(df['year'])
plt.show()
```

```python
f = plt.figure(figsize=(12,4))

f.add_subplot(1,2,1)
sns.countplot(df['average_price'])

f.add_subplot(1,2,2)
plt.boxplot(df['average_price'])
plt.show()
```

```python
f = plt.figure(figsize=(12,4))

f.add_subplot(1,2,1)
df['average_price'].plot(kind='kde')

f.add_subplot(1,2,2)
plt.boxplot(df['average_price'])
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.pairplot(data=df, x_vars=["average_price","total_volume","type","year","geography"], y_vars=['average_price'], size=5, aspect=0.80)
plt.show()
```
